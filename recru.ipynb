```python
%sql

-- view 1
CREATE OR REPLACE TEMP VIEW base AS
SELECT user_id, issue_id, status, updated_at
FROM issues;

-- view 2
CREATE OR REPLACE TEMP VIEW ranked AS
SELECT *,
    -- row number po user as rn  
    FROM base;

-- view 3
CREATE OR REPLACE TEMP VIEW latest AS
SELECT *
FROM ranked
WHERE rn = 1;

-- view 4
CREATE OR REPLACE TEMP VIEW counts AS
SELECT user_id, COUNT(*) as issue_count
FROM base
GROUP BY user_id;

-- final
SELECT l.user_id, l.issue_id, l.status, c.issue_count
FROM latest l
JOIN counts c ON l.user_id = c.user_id;
```

```python

import requests

# ---  issues ---
jira_url = "https://mycompany.atlassian.net"
email = "admin@mycompany.com"
token = "ABC123secrettoken"
project_key = "DE"

auth = (email, token)
headers = {"Accept": "application/json"}

response = requests.get(
    f"{jira_url}/rest/api/3/search",
    auth=auth,
    headers=headers,
    params={"jql": f"project={project_key}", "maxResults": 50}
)
issues = response.json()["issues"]
print(f"Found {len(issues)} issues")


```
